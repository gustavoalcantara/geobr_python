# Nova seção

```python
import geobr
import pandas as pd
import numpy as np
```

```python
micro = geobr.read_micro_region(year=2020)
```

```python
micro
```

```python
micro['code_state'] = micro['code_state'].astype(np.int64)
micro['code_micro'] = micro['code_micro'].astype(np.int64)
```

```python
micro
```

```python
micro=micro.rename({'code_state':'id_uf', 'abbrev_state':'sigla_uf', 'code_micro':'id_microrregiao','geometry':'geometria', 'name_micro': 'nome_microrregiao'},axis='columns')
```

```python
micro=micro.filter(items=['id_uf', 'sigla_uf', 'id_microrregiao', 'geometria'])
```

```python
micro
```

```python
from posixpath import sep
```

```python
micro.to_csv('microrregiao.csv', sep=',', index=False, header=True)
```
